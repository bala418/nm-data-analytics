```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv('../dataset/insurance.csv')
df2.head()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=df2['charges'],y=df2['smoker'])
plt.show()
```

```python
sns.scatterplot(x=df2['charges'],y=df2['smoker'])
```

```python
sns.scatterplot(x=df2['charges'],y=df2['region'])
```

```python
sns.scatterplot(x=df2['bmi'],y=df2['charges'])
```

```python
# create a new column which categorizes the charges

df2['chargescat']= df2['charges'].apply(lambda x: 'low' if x<10000 else 'medium' if x<20000 else 'high')
```

```python
sns.jointplot(x=df2['age'],y=df2['bmi'],hue='chargescat',palette='bright',data=df2)
plt.show()
```

```python
sns.jointplot(data = df2, x = 'charges', y ='bmi');
plt.show() 
```

```python
sns.jointplot(data = df2, x = 'charges', y ='bmi', kind='reg');
plt.show() 
```

```python
sns.scatterplot(x=df2.index,y=df2['age'],hue = df2['chargescat'])
```
